```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Extremes.csv")
print(data)
```

```python
max = data["Maximum"]
print(max)
min = data["Minimum"]
print(min)
ave = data["Average"]
print(ave)
```

```python
from scipy.stats import genextreme

mu, sigma, xi = genextreme.fit(min)
print("Estimated parameters (mu, sigma, xi):", sigma, xi, mu)

mins = np.linspace(-20, 0, 10000)
minextreme= genextreme.pdf(mins, mu, sigma, xi)
plt.plot(mins,minextreme, color = 'b')
plt.xlim(-20,0)
plt.ylim(0,0.18)
plt.xlabel("Annual Minima (Degrees Celcius)")
plt.ylabel("Probability Density")
plt.title("GEV Fitted To Annual CET Minimums 1878-2023")
```

```python
from scipy.stats import genextreme

mu, sigma, xi = genextreme.fit(max)
print("Estimated parameters (mu, sigma, xi):", sigma, xi, mu)

maxes = np.linspace(0, 40, 100000)
maxextreme = genextreme.pdf(maxes, mu, sigma, xi)
plt.plot(maxes,maxextreme, color = 'red')
plt.xlim(20,40)
plt.ylim(0,0.18)
plt.xlabel("Annual Maxima (Degrees Celcius)")
plt.ylabel("Probability Density")
plt.title("GEV Fitted To Annual CET Maximums 1878-2023")
```
